```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import fftpack
```

```python
import csv
# 下面是按照列属性读取的
d = pd.read_csv("sunspot_data.csv", usecols=['Year', 'Number of Sunspots'])
d1 = d.groupby(['Year']).agg({"Number of Sunspots":"sum"})
d1.head()

d1.plot(color="y",linestyle='-',figsize=(12,8))

nos = np.array(d1['Number of Sunspots'])

power = fftpack.fft(nos)
frequencies = fftpack.fftfreq(len(nos), 1)
periods = 1 / frequencies
plt.figure(figsize=(12,8))
plt.plot(periods, abs(power) * 1e-6, '.')
plt.xlabel('Period')
plt.ylabel('Power ($\cdot10^6$)')
plt.xlim(0, 20)
```

方波图形

```python
x = np.linspace(-4*np.pi,4*np.pi,3600)
y = 0
j = [1,5,10,20,50,100,1000,10000,100000]
for k in range(9):
    y = 0
    for i in range(1,j[k]+1,2):
        y += np.sin(i*x)/i
    plt.figure()
    plt.plot(x,y)
```
